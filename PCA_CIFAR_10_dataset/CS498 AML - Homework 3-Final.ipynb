```python
def generateMeanEigvec(dataset):
    dataset = np.array(dataset)
    cov = np.cov(dataset,rowvar=False)
    mean = np.mean(dataset,axis=0)
    eigval,eigvec = np.linalg.eig(cov)
    arg = np.argsort(eigval)
    sorteigval = np.flip(eigval[arg])
    sorteigvec = np.flip(eigvec[:,arg],axis=1)
    return mean,sorteigvec,sorteigval
    
def generateNew(dataset,mean,eigvec,pc):
    dataset = np.array(dataset)
    datahat = np.zeros((np.shape(dataset)[0],np.shape(dataset)[1]))
    for idx, row in enumerate(dataset,0):
        sub = dataset[idx]-mean
        datahat[idx] = mean
        for j in range(0,pc):
            summation =np.dot(eigvec[:,j],sub)
            datahat[idx] = datahat[idx]+summation*eigvec[:,j]
    return datahat
w
```
